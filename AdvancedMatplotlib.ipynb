```python
# 3D surface plot 
```

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
x,y
```

```python
xx,yy = np.meshgrid(x,y)
z = xx**2 + yy**2
```

```python
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection = '3d')
var = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(var)

```

```python
z = np.sin(xx) + np.cos(yy)
```

```python
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection = '3d')
var = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(var)


```

```python
#CONTOUR PLOT 
```

```python
fig = plt.figure(figsize=(12,8))
ax = plt.subplot()
var = ax.contour(xx,yy,z)


```

```python
fig = plt.figure(figsize=(12,8))
ax = plt.subplot()
var = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(var)



```
