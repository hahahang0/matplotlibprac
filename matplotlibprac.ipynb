```python
# from a pandas dataframe 
batsman = pd.read_csv('sharma-kohli.csv')
```

```python
batsman
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
#labels title
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title("Rohit Sharma VS Virat Kohli Career Comparison")
plt.xlabel("IPL Season")
plt.ylabel("Runs Scored")
```

```python
#COLORS(HEX) and Line(WIDTH AND STYLE) and marker(SIZE)

plt.plot(batsman['index'],batsman['V Kohli'],color="green")
plt.plot(batsman['index'],batsman['RG Sharma'],color="black")
plt.title("Rohit Sharma VS Virat Kohli Career Comparison")
plt.xlabel("IPL Season")
plt.ylabel("Runs Scored")
```

```python
#LINE STYLES and LINE WIDTH 

plt.plot(batsman['index'],batsman['V Kohli'],color="green",linestyle="dotted",linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color="black",linestyle="dashed",linewidth=2)
plt.title("Rohit Sharma VS Virat Kohli Career Comparison")
plt.xlabel("IPL Season")
plt.ylabel("Runs Scored")
```

```python
#MARKERS 

plt.plot(batsman['index'],batsman['V Kohli'],color="green",linestyle="dotted",linewidth=3,marker="D")
plt.plot(batsman['index'],batsman['RG Sharma'],color="black",linestyle="dashed",linewidth=2,marker="*")
plt.title("Rohit Sharma VS Virat Kohli Career Comparison")
plt.xlabel("IPL Season")
plt.ylabel("Runs Scored")
```

```python
#Legend and label 

plt.plot(batsman['index'],batsman['V Kohli'],color="green",linestyle="dotted",linewidth=3,marker="D",label="virat")
plt.plot(batsman['index'],batsman['RG Sharma'],color="black",linestyle="dashed",linewidth=2,marker="*",label="rohit")
plt.title("Rohit Sharma VS Virat Kohli Career Comparison")
plt.xlabel("IPL Season")
plt.ylabel("Runs Scored")
plt.legend()
plt.grid()
```
