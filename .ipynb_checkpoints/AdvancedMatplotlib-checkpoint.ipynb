```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
iris = pd.read_csv('iris.csv')
```

```python
iris
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'])
plt.xlabel("Sapel Length")
plt.ylabel("Petal Length")
```

```python
iris['Species']= iris['Species'].replace({"Iris-setosa":0,'Iris-versicolor':1,'Iris-virginica':2})
```

```python
iris.sample(5)
```

```python
plt.figure(figsize=(15,7))

plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet',alpha=.3)
plt.xlabel("Sapel Length")
plt.ylabel("Petal Length")
plt.colorbar()
```

```python
batters = pd.read_csv('batter.csv')
batters.head(5)
```

```python
batters.shape
```

```python
sample_df = batters.sample(25)
```

```python
sample_df
```

```python
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
x  = [1,2,3,4]
y = [5,6,7,8]
plt.scatter(x,y)
plt.text(1,5,'Point 1')
plt.text(2,6,'Point 2')
plt.text(3,7,'Point 3')
plt.text(4,8,'Point 4')
```

```python
# HORIZONTAL AND VERTICAL LINES 
```

```python
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])
plt.axhline(130,color='red')
plt.axhline(140,color='green')
plt.axvline(30,color='red')
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
# a diff way to plot graphs (SUBPLOTS )
```

```python
batters.head()
```

```python
fig,axes = plt.subplots(figsize=(15,6))
axes.scatter(batters['avg'],batters['strike_rate'])
axes.set_title("Something")
axes.set_xlabel("Avg")
axes.set_ylabel("Strike rate")
```

```python
fig,axes = plt.subplots(2,2,sharex=True)
axes[0,0].scatter(batters['avg'],batters['strike_rate'],color='red')

axes[1,1].scatter(batters['avg'],batters['runs'])
axes[0,1].set_title("Average vs Strike Rate")
axes[1,0].set_ylabel("Strike Rate")

axes[1].set_title("Average vs Runs")
axes[1].set_ylabel("Runs")
axes[1].set_xlabel("Avg")
```
